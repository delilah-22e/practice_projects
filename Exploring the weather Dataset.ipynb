```python
#Loading of the packages to use for the project.
import pandas as pd
import numpy as np
```

```python
#Importing the data we will be using for the project.
file_path= r"C:\DATA\1. Weather Data.csv"
asst_data=pd.read_csv(file_path)
print(asst_data)
```

```python
#Listing top 10 rows and last 10 rows of the df
asst_data.head(10)
asst_data.tail(10)
```

```python
#Q1. Finding all records where the weather was clear
clear_weather=asst_data[asst_data['Weather']=='Clear']
clear_weather.head()
```

```python
#Q2: Finding the number of times the wind speed was exactly 4km/h
windspeed_4=(asst_data['Wind Speed_km/h']==4).sum()
print(windspeed_4)
```

```python
#Q3: Check if there are any NULL values present in the dataframe
null_values=asst_data.isnull().sum()
print(null_values)
```

```python
#Q4: Rename the column"Weather" to Weather_condition"
asst_data.rename(columns={'Weather':'Weather_Conditions'},inplace=True)
asst_data.head()
```

#Q5: What is the mean visibility of the dataset.

```python
#Q5: What is the mean visibility of the dataset.
mean_vis=(asst_data['Visibility_km']).mean()
print(mean_vis)
```

```python
#Q6: Find the number of records where windspeed is greater than 24km/h and visibility is equal to 25km.
ws_numb=(asst_data[(asst_data['Wind Speed_km/h']>24) & (asst_data['Visibility_km']==25)]).count()
print(ws_numb)
```

#Q7: What is the mean value of each column for each weather condition.

```python
#I assigned a variable to a column i need to exclude while calcilating the mean.
exclude_column=['Date/Time']
 #I returned all the other columns i need to calculate the mean.
numerical_columns=asst_data.loc[:,asst_data.columns.difference(exclude_column)]
#I calculated the mean of all numerical column and grouping them by the weather conditions.
mean_column= numerical_columns.groupby('Weather_Conditions').mean() 
mean_column
```

```python
#Q8:Find all instances where the weather is clear and the relative humidity is greater than 50 or visibility is above 40.
quest_eight=asst_data[(asst_data['Weather_Conditions']=='Clear') & 
                       ((asst_data['Rel Hum_%']>50)|
                       (asst_data['Visibility_km']>40))]
quest_eight
```

```python
#Q9: Find the number of weather condition that include snow.
snow_weather=asst_data[asst_data['Weather_Conditions'].str.contains('Snow',case=False)]
snow_weather.count()
```
